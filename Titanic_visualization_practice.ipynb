# Titanic Data Analysis and Visualization

```python
titanic.groupby("class").aggregate([np.mean, np.median])
```

```python
titanic.groupby(["class", "sex"])["age"].mean().unstack()
```

```python
sns.catplot(x = "sex", y =  "age",
           hue = "class", kind = "bar",
           data = titanic)
```

```python
sns.catplot(x = "who", y =  "age",
           hue = "class", kind = "bar",
           data = titanic)
```
